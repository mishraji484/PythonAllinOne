### INT to multiple data type conversion

```python
#Int to float 
num1=100
print(num1,type(num1))
num2=float(num1) #converting here
print(num2,type(num2))
```

```python
#Int to Strint
num1=100
print(num1,type(num1))
text=str(num1) #converting here
print(text,type(text))
```

```python
#Int to boolean
num1=100
print(num1,type(num1))
boolean=bool(num1) #converting here
print(boolean,type(boolean))
```

```python
#Int to complex 
num1=100
print(num1,type(num1))
complexed=complex(num1) #converting here
print(complexed,type(complexed))
```

### String to multiple data type conversion

```python
#String to Int
text="123" #if want to convert an string to int then the value inside string("") must be numerical example(1,1.0) else text gives error.
print(text,type(text))
num=int(text) #converting here
print(num,type(num))
```

```python
#String to Int
text="123" #if want to convert an string to float then the value inside string("") must be numerical example(1,1.0) else text gives error.
print(text,type(text))
num=float(text) #converting here
print(num,type(num))
```

```python
#String to Boolean
text="Hello World" #if we convert str into bool then all the input in str become true in bool but only empty("") string will become false.
print(text,type(text))
boolean=bool(text) #converting here
print(boolean,type(boolean))
```

```python
#String to Boolean
text="123.4" #if want to convert str into bool then value inside string must be in numerics wheather it is int or float else text will give error.
print(text,type(text))
complexed=complex(text) #converting here
print(complexed,type(complexed))
```

### Boolean to multiple data type conversion

```python
#Boolean to int
ans1=True
ans2=False
print("Answer1=",ans1,type(ans1))
print("Answer2=",ans2,type(ans2))
ans1_int=int(ans1) #converting here
ans2_int=int(ans2) #converting here
print("Answer1 converted into Int=",ans1_int,type(ans1_int))
print("Answer2 converted into Int=",ans2_int,type(ans2_int))
```

```python
#Boolean to float
ans1=True
ans2=False
print("Answer1=",ans1,type(ans1))
print("Answer2=",ans2,type(ans2))
ans1_float=float(ans1) #converting here
ans2_float=float(ans2) #converting here
print("Answer1 converted into float=",ans1_float,type(ans1_float))
print("Answer2 converted into float=",ans2_float,type(ans2_float))
```

```python
#Boolean to string
ans1=True
ans2=False
print("Answer1=",ans1,type(ans1))
print("Answer2=",ans2,type(ans2))
ans1_str=str(ans1) #converting here
ans2_str=str(ans2) #converting here
print("Answer1 converted into float=",ans1_str,type(ans1_str))
print("Answer2 converted into float=",ans2_str,type(ans2_str))
```

```python
#Boolean to complex
ans1=True
ans2=False
print("Answer1=",ans1,type(ans1))
print("Answer2=",ans2,type(ans2))
ans1_complex=complex(ans1) #converting here
ans2_complex=complex(ans2) #converting here
print("Answer1 converted into float=",ans1_complex,type(ans1_complex))
print("Answer2 converted into float=",ans2_complex,type(ans2_complex))
```

### Complex to multiple data type conversion

```python
#Complex to string
comp=2+10j
print(comp,type(comp))
text=str(comp) #converting here
print(text,type(text))
```

```python
#Complex to boolean
comp=1+10j
print(comp,type(comp))
boolean=bool(comp) #converting here
print(boolean,type(boolean))
```

```python
#Complex to boolean
comp=0+0j
print(comp,type(comp))
boolean=bool(comp) #converting here
print(boolean,type(boolean))
```
