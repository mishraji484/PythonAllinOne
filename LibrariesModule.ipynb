# **Random Module**

```python
#1. random
import random as rd
print("Random:",rd.random())
print("----------------------------------")

#2. randint
print("Randint:",rd.randint(1,10))
print("----------------------------------")

#3. randrange
print("Randrange:",rd.randrange(1,20,2))
print("----------------------------------")

#4. choice
num=[1,2,3,4,5]
print("Choice:",rd.choice(num))
print("----------------------------------")

#choices
print("Choices:",rd.choices(num,k=2))
print("----------------------------------")

#Shuffle
num1=[1,2,3,4,5]
rd.shuffle(num1)
print("Shuffle:",num1)
print("----------------------------------")

#Sample
num2=[1,2,3,4,5]
print("Sample:",rd.sample(num2,3))
print("----------------------------------")

#uniform
print("Uniform:",rd.uniform(10,20))
print("----------------------------------")
```

# **Math Module**

```python
# Ceil
import math
a=10.4
print("Ceil:",math.ceil(a))
print("----------------------------------")


#Floor
print("Floor:",math.floor(a))
print("----------------------------------")


#Sqrt
b=4
print("Sqrt:",math.sqrt(b))
print("----------------------------------")


#pow
print("Pow:",math.pow(10,2))
print("----------------------------------")


#Factorial
print("Factorial:",math.factorial(5))
print("----------------------------------")


#Fabs
print("Fabs:",math.fabs(1))
print("----------------------------------")



#gcd
print("gcd:",math.gcd(28,8))
print("----------------------------------")



#log
print("Log:",math.log(8,2))
print("----------------------------------")



#log10
print("Log10:",math.log10(100))
print("----------------------------------")


#exp
print("Exp:",math.exp(10))  #e=2.718281828459045
print("----------------------------------")


```

# **DateTime Module**

```python
#1. datetime
import datetime as dt
print("Datetime:",dt.datetime.now())
print("----------------------------------")


#2. date
print("Date:",dt.date(2026,3,28))
print("----------------------------------")


#3. time
print("Time:",dt.time(12,10))
print("----------------------------------")


#4. timedelta
print("Timedelta:",dt.timedelta(hours=2,minutes=39,seconds=21))
print("----------------------------------")


#5. now
print("Now:",dt.datetime.now())
print("----------------------------------")


#6. today
print("Today:",dt.date.today())
print("----------------------------------")


#7. strftime
d=dt.datetime.now()
print("strftime:",d.strftime("%d-%m-%y %H:%M:%S"))
print("----------------------------------")


#8. timestamp
now = dt.datetime.now()
ts = now.timestamp()
print("Timestamp:",ts)
print("----------------------------------")


#9. fromtimestamp
ts = 1711539000
print("Fromtimestamp:",dt.datetime.fromtimestamp(ts))
print("----------------------------------")
```
