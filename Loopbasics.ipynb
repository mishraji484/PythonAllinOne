1. Print numbers from 1 to 10 using a for loop.

2. Print numbers from 1 to 20 (only even numbers).

3. Print numbers from 1 to 20 (only odd numbers).

4. Print numbers from 10 to 1 (reverse order).

5. Print the multiplication table of a given number (e.g., 5).

6. Find the sum of numbers from 1 to 100.

7. Find the sum of even numbers from 1 to 50.

8. Print all elements of a list.
   Example list: [10, 20, 30, 40, 50]

9. Count the number of elements in a list without using len().

10. Find the largest number in a list.

11. Count how many numbers between 1 to 100 are divisible by 3.

12. Print all numbers between 1 to 50 that are divisible by both 3 and 5

```python
#1. Print numbers from 1 to 10 using a for loop.
for i in range(1,11):
    print(i,end="--")
    
```

```python
#2. Print numbers from 1 to 20 (only even numbers).
for i in range(2,21,2):
    print(i,end="--")
```

```python
#3. Print numbers from 1 to 20 (only odd numbers).
for i in range(1,20,2):
    print(i,end="--")
```

```python
#4. Print numbers from 10 to 1 (reverse order).
for i in range(10,0,-1):
    print(i,end="--")
```

```python
#5. Print the multiplication table of a given number (e.g., 5).
for i in range(0,51,5):
    print(i,end="--")
```

```python
#6. Find the sum of numbers from 1 to 100.
sum=0
for i in range(1,101):
    sum+=i
print("Sum of all the numbers from 1 to 100 is",sum)
```

```python
#7. Find the sum of even numbers from 1 to 50.
sum=0
for i in range(1,50,2):
    sum+=i
print()
```

```python
#8. Print all elements of a list.
   #Example list: [10, 20, 30, 40, 50]
list=[10,20,30,40,50]
for i in list:
    print(i)
```

```python
#9. Count the number of elements in a list without using len().
list=[10,20,30,40,50]
count=0
for i in list:
    count+=1
print("There are total of",count,"elements in given list.")
```

```python
#10. Find the largest number in a list.
list=[10,20,40,15,11]
largest=list[0]
for i in list:
    if i>largest:
        largest=i
print("Largest number from given list is",largest)        
        

```

```python
#11. Count how many numbers between 1 to 100 are divisible by 3.
count=0
for i in range(1,100):
    if i%3==0:
        count+=1
print(count,"numbers are divisible by 3 between 1 to 100.")
```

```python
#12. Print all numbers between 1 to 50 that are divisible by both 3 and 5
for i in range(1,51):
    if i%3==0 and i%5==0:
        print(i)
```
