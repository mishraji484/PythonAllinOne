```python
#1. Write a program to take a user's name as input and print a greeting message.
id=input("Enter the username:")
print(f"Happy birthday {id}, many many happy returns of the day.")
```

```python
#2. Take two integers as input from the user and print their sum.
num1=int(input("Enter the first numbers="))
num2=int(input("Enter the Second numbers="))
sum=num1+num2
print(f"Sum of {num1} and {num2} is {sum}")
```

```python
#3. Take a float input from the user and print its square.
num1=float(input("Enter the number:"))
print(f"Square of {num1} is {num1**2}")
```

```python
#4. Take three numbers as input and print the largest among them.
a=eval(input("Enter the first no.:"))
b=eval(input("Enter the second no.:"))
c=eval(input("Enter the third no.:"))
if a>b and a>c:
    print("A is largest:",a)
elif b>a and b>c:
    print("B is largest:",b)
else:
    print("C is largest:",c)
```

```python
#5. Take a string input and print its length.
text="Raghvendra Kumar Mishra"
count=0
for i in text:
    if i in text:
        count+=1
print(f"Length of {text} is",count)
```

```python
#7. Take a number as input and check whether it is even or odd.
num=eval(input("Enter the number:"))
if num%2==0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
#8. Take a character input and check whether it is a vowel or consonant.
word=input("Enter the character:")
vowel="AEIOUaeiou"
if word in vowel:
    print(f"{word} is an vowel.")
else:
    print(f"{word} is a consonent.")
```
