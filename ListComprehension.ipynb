```python
'''1. Create a list of numbers from 1 to 10 using list comprehension
   Output: [1, 2, 3, ..., 10]'''
list1=[i for i in range(1,11)]
print(list1)
```

```python
'''2. Create a list of squares of numbers from 1 to 5
   Output: [1, 4, 9, 16, 25]'''
list2=[i**2 for i in range(1,6)]
list2
```

```python
'''3. Create a list of even numbers from 1 to 10
   Output: [2, 4, 6, 8, 10]'''
list3=[i for i in range(1,11) if i%2==0]
list3
```

```python
'''4. Create a list of odd numbers from 1 to 10
   Output: [1, 3, 5, 7, 9]'''
list4=[i for i in range(1,11) if i%2!=0]
list4
```

```python
'''5. Convert all characters of a string to a list
   Input: "hello"
   Output: ['h', 'e', 'l', 'l', 'o']
'''
string="hello"
list5=[i for i in string]
list5
```

```python
'''6. Create a list of squares of even numbers from 1 to 10
   Output: [4, 16, 36, 64, 100]'''
list6=[i**2 for i in range(1,11) if i%2==0]
list6
```

```python
'''7. Create a list of numbers divisible by 3 from 1 to 20
   Output: [3, 6, 9, 12, 15, 18]'''
list7=[i for i in range(1,20) if i%3==0]
list7
```

```python
'''8. Convert a list of strings to uppercase
   Input: ["a", "b", "c"]
   Output: ["A", "B", "C"]'''
input=['a','b','c']
list8=[i.upper() for i in input]
list8
```

```python
'''9. Extract only vowels from a string
   Input: "education"
   Output: ['e', 'u', 'a', 'i', 'o']
'''
Input= "education"
vowels='aeiouAEIOU'
list9=[i for i in Input if i in vowels]
list9
```

```python
'''10. Replace negative numbers with 0 in a list
    Input: [1, -2, 3, -4, 5]
    Output: [1, 0, 3, 0, 5]'''
Input= [1, -2, 3, -4, 5]    
list10=[0 if i<0 else i for i in Input]
list10
```

```python
'''11. Create a list where even numbers are squared and odd numbers remain same
    Input: [1, 2, 3, 4]
    Output: [1, 4, 3, 16]'''
input=[1,2,3,4]
list11=[i**2 if i%2==0 else i for i in input]
print(list11)
```

```python
'''12. Create a list with "Even" or "Odd" labels
    Input: [1, 2, 3, 4]
    Output: ["Odd", "Even", "Odd", "Even"]'''
input=[1,2,3,4]
list12=["even" if i%2==0 else "odd" for i in input]
print(list12)   
```

```python
'''13. Filter only positive numbers
    Input: [-1, 2, -3, 4]
    Output: [2, 4]'''
input=[-1,2,-3,4]
list13=[i for i in input if i>0]
print(list13)
```

```python
'''14. Filter strings with length greater than 3
    Input: ["hi", "hello", "hey"]
    Output: ["hello"]'''
Input= ["hi", "hello", "hey"]
list14=[i for i in Input if len(i)>3 ]
print(list14)
```

```python
'''15. Get ASCII values of characters
    Input: "abc"
    Output: [97, 98, 99]'''
Input= "abc"
list15=[ord(i) for i in Input]
print(list15)
```

```python
'''16. Flatten a 2D list
    Input: [[1, 2], [3, 4], [5]]
    Output: [1, 2, 3, 4, 5]'''
Input= [[1, 2], [3, 4], [5]]
list16= [j for i in Input for j in i]
list16
```

```python
'''17. Create a multiplication table (1 to 3)
    Output: [1, 2, 3, 2, 4, 6, 3, 6, 9]
'''
list17=[i*j for i in range(1,4) for j in range(1,4)]
list17
```

```python
'''18. Create pairs from two lists
    Input: [1, 2], ['a', 'b']
    Output: [(1, 'a'), (1, 'b'), (2, 'a'), (2, 'b')]'''
list1 = [1, 2]
list2 = ['a', 'b']

pairs = [(i, j) for i in list1 for j in list2]
pairs
```

```python
'''19. Flatten list but keep only even numbers
    Input: [[1, 2], [3, 4], [5, 6]]
    Output: [2, 4, 6]'''
Input= [[1, 2], [3, 4], [5, 6]]
list16= [j for i in Input for j in i if j%2==0]
list16
```

```python
'''20. Remove duplicates using list comprehension
    Input: [1, 2, 2, 3, 1]
    Output: [1, 2, 3]'''
Input= [1, 2, 2, 3, 1]
result=[]
list20=[result.append(i) for i in Input if i not in result]
result
```

```python
'''21. Find common elements between two lists
    Input: [1, 2, 3], [2, 3, 4]
    Output: [2, 3]'''
lis1= [1, 2, 3]
lis2= [2, 3, 4]   
list21=[i for i in lis1 for j in lis2 if i==j]
print(list21)
```

```python
'''22. Create list of tuples (number, square)
    Input: [1, 2, 3]
    Output: [(1, 1), (2, 4), (3, 9)]'''
Input= [1, 2, 3]
list22=[(i,i**2) for i in Input]
list22
```

```python
'''23. Reverse each string in a list
    Input: ["abc", "def"]
    Output: ["cba", "fed"]'''
Input= ["abc", "def"]
list23=[i[::-1] for i in Input]
list23
```

```python
'''24. Extract digits from a string
    Input: "a1b2c3"
    Output: ['1', '2', '3']'''
Input= "a1b2c3"
list24=[i for i in Input if i.isnumeric()]
list24
```

```python
'''25. Create list of prime numbers from 1 to 20
    Output: [2, 3, 5, 7, 11, 13, 17, 19]'''
list25=[i for i in range(2,21) if all(i%j!=0 for j in range(2,i))]
list25
```

```python
'''26. Matrix transpose using list comprehension
    Input: [[1, 2, 3], [4, 5, 6]]
    Output: [[1, 4], [2, 5], [3, 6]]'''
Input= [[1, 2, 3], [4, 5, 6]]
list26=[[j[i] for j in Input] for i in range(len(Input[0]))]
list26
```

```python
'''27. Flatten 3D list
    Input: [[[1, 2], [3]], [[4, 5]]]
    Output: [1, 2, 3, 4, 5]'''
Input= [[[1, 2], [3]], [[4, 5]]]  
list27=[k for i in Input for j in i for k in j]
list27 
```

```python
'''28. Create list of factorial values from 1 to 5
    Output: [1, 2, 6, 24, 120]'''
import math
list28=[math.factorial(i) for i in range(1,6)]
list28
```

```python
'''29. Create list of words starting with vowel
    Input: ["apple", "banana", "orange", "grape"]
    Output: ["apple", "orange"]'''
Input= ["apple", "banana", "orange", "grape"]
vowels='aeiouAEIOU'
list29=[i for i in Input if i[0] in vowels]
list29
```

```python
'''30. Generate list of all substrings of a string
    Input: "abc"
    Output: ["a", "ab", "abc", "b", "bc", "c"]'''
Input= "abc"
list30=[Input[i:j] for i in range(len(Input)) for j in range(i+1, len(Input)+1)]
list30
```
