```python
#1. Print numbers from 1 to 10 using a for loop.
for i in range(1,11):
    print(i,end="--")
    
```

```python
#2. Print numbers from 1 to 20 (only even numbers).
for i in range(2,21,2):
    print(i,end="--")
```

```python
#3. Print numbers from 1 to 20 (only odd numbers).
for i in range(1,20,2):
    print(i,end="--")
```

```python
#4. Print numbers from 10 to 1 (reverse order).
for i in range(10,0,-1):
    print(i,end="--")
```

```python
#5. Print the multiplication table of a given number (e.g., 5).
for i in range(0,51,5):
    print(i,end="--")
```

```python
#6. Find the sum of numbers from 1 to 100.
sum=0
for i in range(1,101):
    sum+=i
print("Sum of all the numbers from 1 to 100 is",sum)
```

```python
#7. Find the sum of even numbers from 1 to 50.
sum=0
for i in range(1,50,2):
    sum+=i
print()
```

```python
#8. Print all elements of a list.
   #Example list: [10, 20, 30, 40, 50]
list=[10,20,30,40,50]
for i in list:
    print(i)
```

```python
#9. Count the number of elements in a list without using len().
list=[10,20,30,40,50]
count=0
for i in list:
    count+=1
print("There are total of",count,"elements in given list.")
```

```python
#10. Find the largest number in a list.
list=[10,20,40,15,11]
largest=list[0]
for i in list:
    if i>largest:
        largest=i
print("Largest number from given list is",largest)        
        

```

```python
#11. Count how many numbers between 1 to 100 are divisible by 3.
count=0
for i in range(1,100):
    if i%3==0:
        count+=1
print(count,"numbers are divisible by 3 between 1 to 100.")
```

```python
#12. Print all numbers between 1 to 50 that are divisible by both 3 and 5
for i in range(1,51):
    if i%3==0 and i%5==0:
        print(i)
```

```python
#13. Find the factorial of a number using a for loop.
fact=1
for i in range(1,6):
    fact*=i
print("Factorial of 5 is",fact)
```

```python
#14. Reverse a given number using a for loop.
    #Example: 1234 → 4321
num=1234
rev=''
for i in str(num):
    rev=i+rev
print(rev)
```

```python
#15. Print each character of a string using a for loop.
text="Hello"
for i in text:
    print(i)
```

```python
#16. Count vowels in a given string.
text="Hello everyone my name is raghvendra kumar mishra"
vowels="AEIOUaeiou"
count=0
for i in text:
    if i in vowels:
        count+=1
print(count)
```

```python
#17. Count how many digits are in a number.
   # Example: 12345 → 5 digits
num=12345
count=0
for i in str(num):
    count+=1
print(count,"digits")
```

```python
#18. Print the square of numbers from 1 to 10.
for i in range(1,11):
    sqr=i**2
    print(sqr)
```

```python
#19. Print the cube of numbers from 1 to 10.
for i in range(1,11):
    sqr=i**3
    print(sqr)
```

```python
#20. Find the sum of digits of a number.
    #Example: 123 → 6
num=123
total=0
for i in str(num):
    total+=int(i)
print(total)
    
```
