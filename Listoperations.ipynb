```python
#23. Find index of element (without using index())
#Input: List = [10, 20, 30], Element = 20
lis=[10,20,30]
for i in range(len(lis)):
    if lis[i]==20:
        print(i)
```

```python
#24. Rotate list to right by 1 position
#Input: [1, 2, 3, 4]
lis=[1, 2, 3, 4]

last =lis[-1]

for i in range(len(lis) - 1, 0, -1):
    lis[i] = lis[i - 1]

lis[0] = last

print(lis)
```

```python
#25. Find common elements in two lists
#Input: [1, 2, 3], [2, 3, 4]
l1=[1,2,3]
l2=[2,3,4]
result=[]
for i in l1:
    if i in l2:
        result.append(i)
print(result)
```

```python
#26. Flatten a nested list (1 level only)
#Input: [1, [2, 3], 4]
lis= [1, [2, 3], 4]

result=[]

for i in lis:
    if type(i)==list:
        for j in i:
            result.append(j)
    else:
        result.append(i)

print(result)
```
