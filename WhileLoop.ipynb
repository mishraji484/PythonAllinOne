4. Print odd numbers between 1 and 20.

6. Print numbers from 1 to 50 that are divisible by both 3 and 5.
7. Print numbers from 1 to 100 that are not divisible by 3.
8. Print numbers between 1 and 100 that are divisible by 7.

12. Print numbers from 1 to 100 divisible by 3 but not by 5.
13. Print numbers from 1 to 100 divisible by 5 but not by 3.

15. Print numbers from 1 to 100 divisible by both 2 and 3.

```python
#1. Print numbers from 1 to 10 using a while loop.
i=1
while i<11:
    print(i,end="  ")
    i+=1
```

```python
#2. Print numbers from 10 to 1 using a while loop.
i=10
while i>0:
    print(i,end="  ")
    i-=1
```

```python
#3. Print even numbers between 1 and 20 using a while loop and condition.
i=1
while i<21:
    if i%2==0:
        print(i,end="  ")
    i+=1
```

```python
#4. Print odd numbers between 1 and 20.
i=1
while i<20:
    print(i,end="  ")
    i+=2
```

```python
#5. Print numbers from 1 to 50 that are divisible by 5.
i=1
while i<51:
    if i%5==0:
        print(i,end="  ")
    i+=1
```

```python
#6. Print numbers from 1 to 50 that are divisible by both 3 and 5.
i=1
while i<51:
    if i%3==0 and i%5==0:
        print(i,"is divisible by both 3 and 5.")
    i+=1
```

```python
#7. Print numbers from 1 to 100 that are not divisible by 3.
i=1
while i<100:
    if i%3!=0:
        print(i,end=" ")
    i+=1
```

```python
#8. Print numbers between 1 and 100 that are divisible by 7.
i=1
while i<100:
    if i%7==0:
        print(i,end=" ")
    i+=1
```

```python
#9. Print numbers from 1 to 30 but skip multiples of 3.
i=1
while i<31:
    if i%3!=0:
        print(i)
    i+=1
    
```

```python
#10. Print numbers from 1 to 20 but stop when the number becomes 15.
i=1
while i<20:
    if i==16:
        break
    print(i,end="  ")
    i+=1
```

```python
#11. Print numbers from 1 to 50 and label them as even or odd.
i=1
while i<51:
    if i%2==0:
        print(i,"is Even.")
    else:
        print(i,"is Odd.")
    i+=1
```

```python
#12. Print numbers from 1 to 100 divisible by 3 but not by 5.
i=1
while i<101:
    if i%3==0 and i%5!=0:
        print(i,end="  ")
    i+=1
        
```

```python
#13. Print numbers from 1 to 100 divisible by 5 but not by 3.
i=1
while i<101:
    if i%5==0 and i%3!=0:
        print(i,end="  ")
    i+=1
```

```python
#14. Print numbers from 1 to 100 divisible by either 3 or 5.
i=1
while i<101:
    if i%5==0 or i%3==0:
        print(i,end="  ")
    i+=1
```

```python
#15. Print numbers from 1 to 100 divisible by both 2 and 3.
i=1
while i<101:
    if i%2==0 or i%3==0:
        print(i,end="  ")
    i+=1
```
