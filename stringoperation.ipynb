```python
#Count occurrence of a character
#Input: String = "banana", Character = 'a'
text="banana"
count=0
for i in text:
    if "a" == i:
        count+=1
print(count)
```

```python
#Print characters at even index positions
#Input: "abcdef"
text="abcdef"
for i in range(len(text)):
    if i%2==0:
        print(text[i],end=" ")
```

```python
#Print characters at odd index positions
#Input: "abcdef"
for i in range(len(text)):
    if i%2!=0:
        print(text[i],end=" ")
```

```python
#Reverse a string (without slicing)
#Input: "hello"
text="hello"
rev=""
for i in text:
    rev=i+rev
print("Reversed:",rev)
```

```python
#Count vowels
#Input: "education"
text="education"
vowels="aeiouAEIOU"
count=0
for i in text:
    if i in vowels:
        count+=1
print("vowels are",count)
```

```python
#Count consonants
# Input: "hello"
text="hello"
vowels="aeiouAEIOU"
count=0
for i in text:
    if i not in vowels:
        count+=1
print("vowels are",count)
```

```python
#Count digits in a string
#Input: "a1b2c3"
string="a1b2c3"
count=0
for i in string:
    if i.isdigit():
        count+=1
print(count)
```

```python
#Count spaces in a string
#Input: "hi hello how"
text="hi hello how"
count=0
for i in text:
    if i==" ":
        count+=1
print("Count of spaces are",count)
```

```python
#Replace spaces with '-'
#Input: "hi hello"
text="hi hello"
result=""
for i in text:
    if i==" ":
        result += "-"
    else:
        result += i

print(result)
```

```python
#Check if two strings are equal (without using ==)
#Input: "hello", "hello"
s1="hello"
s2="hello"
equal=True

for i in range(len(s1)):
    if s1[i] != s2[i]:
        equal=False
if equal:
    print("Equal")
else:
    print("not equal")
    
```

```python
#Merge two strings alternately
#Input: "abc", "123"
s1="abc"
s2="123"
result=""
for i in range(len(s1)):
    result+=s1[i]+s2[i]
print(result)
```

```python
#Check if string contains only digits
#Input: "12345"
string="12345"
check=True
for i in string:
    if not i.isdigit():
        check=False
if check:
    print("All are digits.")
else:
    print("all are not digits")
```

```python
#Check if string contains only alphabets
#Input: "hello"
string="hello"
check=True
for i in string:
    if not i.isalpha():
        check=False
if check:
    print("All are alphabets.")
else:
    print("all are not alphabets.")
```

```python
#Find longest repeating character sequence
#Input: "aaabbccccd"
text = "aaabbccccd"

max_count = 1
count = 1
max_char = text[0]

for i in range(len(text) - 1):
    if text[i] == text[i + 1]:
        count += 1
    else:
        if count > max_count:
            max_count = count
            max_char = text[i]
        count = 1

# check last group
if count > max_count:
    max_count = count
    max_char = text[-1]

print(max_char, max_count)
```

```python
#Shift each character by 1
#Input: "abc"
text="abc"
result=""
for i in text:
    result += chr(ord(i) + 1)
print(result)
```
