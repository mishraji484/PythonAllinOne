```python
'''1. Print all elements of a tuple using a for loop
   Input: (1, 2, 3, 4)
   Output: 1 2 3 4'''
tup=(1,2,3,4)
for i in tup:
    print(i, end="  ")
```

```python
'''2. Print all elements using while loop
   Input: (5, 6, 7)
   Output: 5 6 7'''
tup=(5,6,7)
i = 0
while i < len(tup):
    print(tup[i], end=" ")
    i += 1
```

```python
'''
3. Find length of tuple (without using len)
   Input: (10, 20, 30)
   Output: 3'''
tup= (10,20,30)
count=0
for i in tup:
    count+=1
print(count)
```

```python
'''4. Find sum of all elements
   Input: (1, 2, 3, 4)
   Output: 10'''
tup=(1,2,3,4)
sum=0
for i in tup:
    sum+=i
print(sum)
```

```python
'''5. Find maximum element
   Input: (3, 7, 2, 9, 5)
   Output: 9'''
tup=(3,7,2,9,5)
maximum=tup[0]
for i in tup:
    if i>maximum:
        maximum=i
print(maximum)
```

```python
'''6. Find minimum element
   Input: (3, 7, 2, 9, 5)
   Output: 2'''
tup=(3,7,2,9,5)
minimum=tup[0]
for i in tup:
    if i<minimum:
        minimum=i
print(minimum)
```

```python
'''7. Count occurrence of a number
   Input: Tuple = (1, 2, 2, 3, 2), Number = 2
   Output: 3'''
tup=(1,2,2,3,2)
count=0
for i in tup:
    if i==2:
        count+=1
print(count)
```

```python
'''8. Create a new tuple by adding an element
   Input: (1, 2, 3), Add = 4
   Output: (1, 2, 3, 4)'''
tup=(1,2,3)
tup=list(tup)
tup.append((4))
tup=tuple(tup)
print(tup)
```

```python
'''9. Insert an element at specific index
   Input: Tuple = (1, 2, 4), Element = 3, Index = 2
   Output: (1, 2, 3, 4)'''
tup=(1,2,4)
tup=list(tup)
tup.insert(2,3)
tup=tuple(tup)
print(tup)
```

```python
'''10. Remove first occurrence of an element
    Input: (1, 2, 3, 2), Remove = 2
    Output: (1, 3, 2)'''
tup=(1,2,3,2)
tup=list(tup)
tup.remove(2)
tup=tuple(tup)
print(tup)
```

```python
'''11. Remove last element
    Input: (10, 20, 30, 40)
    Output: (10, 20, 30)'''
tup=(10,20,30,40)
tup=list(tup)
tup.pop()
tup=tuple(tup)
print(tup)
```

```python
'''12. Reverse a tuple
    Input: (1, 2, 3)
    Output: (3, 2, 1)'''
tup=(1,2,3)
tup=list(tup)
tup=tup[::-1]
tup=tuple(tup)
print(tup)
```

```python
'''13. Copy a tuple
    Input: (1, 2, 3)
    Output: (1, 2, 3)'''
tup=(1,2,3)
temp=tup
print(temp)
```

```python
'''14. Merge two tuples
    Input: (1, 2), (3, 4)
    Output: (1, 2, 3, 4)'''
t1=(1,2)
t2=(3,4)
t1+t2
```

```python
'''15. Remove duplicates from tuple
    Input: (1, 2, 2, 3, 1)
    Output: (1, 2, 3)'''
tup=(1,2,2,3,1)
temp=list(tup)
result=[]
for i in temp:
    if i not in result:
        result.append(i)
result=tuple(result)
print(result)
```

```python
'''16. Sort tuple in ascending order
    Input: (5, 2, 8, 1)
    Output: (1, 2, 5, 8)'''
tup=(5,2,8,1)
temp=list(tup)
temp.sort()
tup=tuple(temp)
print(tup)
```

```python
'''17. Sort tuple in descending order
    Input: (5, 2, 8, 1)
    Output: (8, 5, 2, 1)'''
tup=(5,2,8,1)
temp=list(tup)
temp.sort(reverse=True)
tup=tuple(temp)
print(tup)
```

```python
'''18. Find second largest element
    Input: (10, 20, 5, 30)
    Output: 20'''
tup=(10, 20, 5, 30)
largest=tup[0]
for i in tup:
    if i>largest:
        secondlargest=largest
        largest=i
print(secondlargest)
```

```python
'''19. Check if element exists
    Input: Tuple = (1, 2, 3), Element = 2
    Output: Found'''
tup= (1, 2, 3)
for i in tup:
    if i==2:
        print("Found")
```

```python
'''20. Count even numbers
    Input: (1, 2, 3, 4, 5, 6)
    Output: 3'''
tup= (1,2,3,4,5,6)
count=0
for i in tup:
    if i%2==0:
        count+=1
print(count)
```

```python
'''21. Count odd numbers
    Input: (1, 2, 3, 4, 5)
    Output: 3'''
tup=(1,2,3,4,5)
count=0
for i in tup:
    if i%2!=0:
        count+=1
print(count)
```

```python
'''22. Create new tuple with squares of elements
    Input: (1, 2, 3)
    Output: (1, 4, 9)'''
tup=(1, 2, 3)
for i in tup:
    print(i**2,end=" ")
```

```python
'''23. Remove all negative numbers
    Input: (1, -2, 3, -4, 5)
    Output: (1, 3, 5)'''
tup=(1, -2, 3, -4, 5)
for i in tup:
    if i<0:
        continue
    print(i,end="  ")
```

```python
'''24. Find index of an element (without using index())
    Input: Tuple = (10, 20, 30), Element = 20
    Output: 1'''
tup = (10, 20, 30)
count=0
for i in tup:
    if i==20:
        count+=1
print(count)
        
```

```python
'''25. Rotate tuple right by 1 position
    Input: (1, 2, 3, 4)
    Output: (4, 1, 2, 3)'''
tup=(1,2,3,4)


rotated = (tup[-1],) + tup[:-1]
print(rotated)
```

```python
'''26. Find common elements between two tuples
    Input: (1, 2, 3), (2, 3, 4)
    Output: (2, 3)'''
t1=(1,2,3)
t2=(2,3,4)
for i in range(len(t1)):
    if t1[i] in t2:
        print(t1[i],end="  ")
```

```python
'''27. Flatten a nested tuple (1 level)
    Input: (1, (2, 3), 4)
    Output: (1, 2, 3, 4)'''
tup = (1, (2, 3), 4)

result = []
for i in tup:
    if isinstance(i, tuple):
        for j in i:
            result.append(j)
    else:
        result.append(i)

result = tuple(result)
print(result)
```

```python
'''28. Remove element at specific index
    Input: Tuple = (1, 2, 3, 4), Index = 2
    Output: (1, 2, 4)'''
tup= (1, 2, 3, 4)
temp=list(tup)
temp.pop(2)
tup=tuple(temp)
print(tup)
```

```python
'''29. Count frequency of each element
    Input: (1, 2, 2, 3, 1)
    Output: {1:2, 2:2, 3:1}'''
tup=(1,2,2,3,1)
freq = {}
for i in tup:
    freq[i] = freq.get(i, 0) + 1

print(freq)
```

```python
'''30. Check if tuple is palindrome
    Input: (1, 2, 3, 2, 1)
    Output: Palindrome'''
tup = (1, 2, 3, 2, 1)

if tup == tup[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
