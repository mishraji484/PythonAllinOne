## Mixed Questions

```python
"""1. Take a number and print:
   - Its type
   - Binary
   - Octal
   - Hexadecimal"""
num=45
print(f"Type of {num} =",type(num))
print(f"{num} into binary =",bin(num))
print(f"{num} into octal =",oct(num))
print(f"{num} into Hexadecimal =",hex(num))
```

```python
'''2. Print:
   Decimal: 10
   Binary: ?
   Octal: ?
   Hex: ?'''
decimal=10
print("Binary=",bin(decimal))
print("Octal=",oct(decimal))
print("Hexadecimal=",hex(decimal))
```

```python
#3. Convert int → binary → back to int and check type.
num=24
binary=bin(num)
print("Int to binary=",binary)
num1=int(binary,2)
print("Binary to int=",num1)
print(type(num1))
```
