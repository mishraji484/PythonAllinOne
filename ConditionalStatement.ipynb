## **IF Statement**

```python
#2. Check if a number is greater than 10.
num=int(input("Enter a number:"))
if num>10:
    print(num,"is greater than 10.")
print("Executed.")
```

```python
#3. Check if a number is even.
num=int(input("Enter a number:"))
if num%2==0:
    print(num,"is even.")
print("Executed.")
```

```python
#4. Check if a number is divisible by 5.
num=int(input("Enter a number:"))
if num%5==0:
    print(num,"is divisible by 5.")
print("Executed.")
```

```python
#5. Check if a person is eligible to vote (age ≥ 18).
age=int(input("Enter the age:"))
if age>=18 and age<100:
    print("You are eligible to vote.")
print("Executed.")
```

```python
#7. Check if a number is less than 0.
num=float(input("Enter a number:"))
if num<0:
    print(num,"is less than 0.")
print("Executed.")
```

```python
#8. Check if a number is equal to 100.
num=int(input("Enter a number:"))
if num==100:
    print("Given number is equal to 100.")
print("Executed.")   
```

##  **If-Elif-Else Statements**

```python
#21. Check if a number is positive, negative, or zero.
num=float(input("Enter a number:"))
if num>0:
    print("Given number is positive.")
elif num==0:
    print("Given number is zero.")
else:
    print("Given number is negative.")
```

```python
#22. Check if a number is divisible by 2, 3, or neither.
num=float(input("Enter a number:"))
if num%2==0:
    print(num,"is divisible by 2.")
elif num%3==0:
    print(num,"is divisible by 3.")
else:
    print(num,"is neither divisible by 2 and 3.")
```

```python
#23. Check the grade based on marks:
'''Marks ≥ 90 → A
Marks ≥ 75 → B
Else → C'''
marks=float(input("Enter your marks:"))
if marks>=90 and marks<=100:
    print("A grade")
elif marks>=75 and marks<90:
    print("B grade")
else:
    print("C grade")
```

```python
#24. Check age category:
'''Age < 13 → Child
Age < 20 → Teen
Else → Adult'''
age=int(input("Enter the age:"))
if age>0 and age<13:
    print("Child.")
elif age>=13 and age<20:
    print("Teen.")
else:
    print("Adult.")
```

```python
#25. Check if a number is small, medium, or large:
'''< 10 → Small
< 50 → Medium
Else → Large'''
num=float(input("Enter a number:"))
if num<10:
    print("Small.")
elif num>=10 and num<50:
    print("Medium.")
else:
    print("Large.")
```

```python
#26. Check the day type based on input (1–7):
#1–5 → Weekday
#6–7 → Weekend
a=int(input("Enter a number to check day type:"))
if a>=1 and a<=5:
    print("Weekday.")
elif a>=6 and a<=7:
    print("Weekend.")
else:
    print("Invalid Input.")
```

```python
#27. Check temperature category:
'''< 15 → Cold
15–30 → Normal
> 30 → Hot'''
temp=float(input("Enter the temperature:"))
if temp<15:
    print("Cold.")
elif temp>=15 and temp<=30:
    print("Normal.")
else:
    print("Hot.")
```

```python
#28. Check salary category:
'''< 20,000 → Low
< 50,000 → Medium
Else → High'''
salary=float(input("Enter the salary="))
if salary<=20000:
    print("Salary is low.")
elif salary>20000 and salary<=50000:
    print("Salary is medium.")
else:
    print("Salary is high.")
```

```python
#29. Check if a number is divisible by 2, 5, or neither.
num=float(input("Enter a number:"))
if num%2==0:
    print(num,"is divisible by 2.")
elif num%5==0:
    print(num,"is divisible by 5.")
else:
    print(num,"is neither divisible by 2 and 5.")
```

```python
#30. Check login status:
#Correct password → Login Successful
#Wrong password → Try Again
stored_password="abc1234"
password=input("Enter the password=")
if password==stored_password:
    print("Login successful")
else:
    print("Try again")
```
